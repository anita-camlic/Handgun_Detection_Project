## Visual Creation

This notebook is for producing all of the visuals used in my presentation.

Here I will create a bar chart of the different validation scores of various models

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
changes = ['Baseline', 'Dense Layers', 'Convolutional Layers', 'Filters', 'Dropout']
scores = [72.14,62.54,78.02,83.90,85.76]
data = {"changes": ['Baseline', 'Dense Layers', 'Convolutional Layers', 'Filters', 'Dropout'],
        "scores": [72.14,62.54,78.02,83.90,88.14]}
```

```python
df = pd.DataFrame(data, columns=['changes', 'scores'])
```

```python
sns.set(rc={'figure.figsize':(11,8)})
plots = sns.barplot(x="changes", y="scores", data=df, palette = 'mako')
plt.xlabel("Iterative Change Made", size=14)
plt.ylabel("Validation Accuracy Score", size=14)

# Iterating over the bars one-by-one
for bar in plots.patches:
   
  # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # free space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment
    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=11, xytext=(0, 8),
                   textcoords='offset points')
plt.xticks(rotation=45)
plt.show();

fig = plots.get_figure()
fig.savefig("iterative_modeling.png",bbox_inches='tight',transparent = True)
```
